```python
#importing libraries
from __future__ import print_function
import numpy as np
import pandas as pd
```

```python
#reading data saved from local directory
df= pd.read_csv('/Users/upasanadhar/Documents/Lakes/ML with Python/datasets/mldata/OGA_Field_Production_PPRS_WGS84.csv',index_col=0)
df.head()
```

```python
#converting into pandas daatframe
df_prod=pd.DataFrame(df)
df_prod
```

```python
#sort with respect to year column
df_prod = df_prod.sort_values('PERIODYR',ascending=True)
df_prod=pd.DataFrame(df_prod)
df_prod
```

```python
#Selecting columns of interest
df_prod_1=df_prod[['PERIODYR','OILPRODMAS','DGASPRODMA','AGASPRODMA','INJWATMASS','WATPRODMAS','SHAPESTArea','SHAPESTLength']]
df_prod_1 = df_prod_1.replace(np.NaN, 0)
df_prod_1.head()
```

```python
df_prod_1.describe().transpose()
```

```python
#Renaming columns
df_prod_1.rename(columns={'OBJECTID':'ID','PERIODYR':'Year','OILPRODMAS':'Oil','DGASPRODMA':'D Gas','AGASPRODMA':'A Gas','INJWATMASS':'Injected water', 'WATPRODMAS':'Water produced','SHAPESTArea':'Area','SHAPESTLength':'Length'},inplace=True)
df_prod_1.head()
```

```python
import numpy as np
#df_prod_1['Oil produced']=np.where(df_prod_1['Oil produced'] > 5000, 1, df_prod_1['Oil produced'],inplace=True)

#df_prod_1['Oil produced']=np.where(df_prod_1['Oil produced'] <= 5000, 0, df_prod_1['Oil produced'],inplace=True)

#df_prod_1.Oil produced.loc[(df_prod_1['Oil produced'] <= 5000)] = 0
#df_prod_1

#df_prod_1['Oil'] = np.where((df_prod_1.Oil > 5000),'1',df_prod_1.Oil)
#df_prod_1

#df_prod_1['Oil'] = np.where((df_prod_1.Oil !< 5000),'0',df_prod_1.Oil)
```

```python
def myfunc(x):
    if x <= 5000:
        return 0
   
    else:
        return 1

df_prod_1['Oil'] = df_prod_1.apply(lambda x: myfunc(x.Oil), axis=1)

df_prod_1
```

```python
X = df_prod_1.drop('Oil',axis=1)
y = df_prod_1['Oil']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Fit only to the training data
scaler.fit(X_train)
```

```python
# Now apply the transformations to the data:
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
#y_train=scaler.transform(y_train)
#y_test=scaler.transform(y_test)
```

```python
#Training the model
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(7,7,5),max_iter=100)
```

```python
mlp.fit(X_train,y_train)
```

```python
#classify as >50,000 oil produced 1 < = 0
```

```python
predictions = mlp.predict(X_test)
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
# STEP 8: CONFUSION MATRIX

# Print out confusion matrix
cmat = confusion_matrix(y_test, predictions)
#print(cmat)
print('TN - True Negative {}'.format(cmat[0,0]))
print('FP - False Positive {}'.format(cmat[0,1]))
print('FN - False Negative {}'.format(cmat[1,0]))
print('TP - True Positive {}'.format(cmat[1,1]))
print('Accuracy Rate: {}'.format(np.divide(np.sum([cmat[0,0],cmat[1,1]]),np.sum(cmat))))
print('Misclassification Rate: {}'.format(np.divide(np.sum([cmat[0,1],cmat[1,0]]),np.sum(cmat))))
```

```python
import matplotlib.pyplot as plt
labels = [0,1] 

cmat = confusion_matrix(y_test, predictions, labels) 

print(cmat) 

fig = plt.figure() 

ax = fig.add_subplot(111) 

cax = ax.matshow(cmat) 

plt.title('Confusion matrix of the classifier') 

fig.colorbar(cax) 

ax.set_xticklabels([''] + labels) 

ax.set_yticklabels([''] + labels) 

plt.xlabel('Predicted') 

plt.ylabel('True') 

plt.show()
```
